```python
# Add dependencies
import glob
import pandas as pd
```

```python
# Combine all the csv data into a new csv

clean_data = pd.concat([pd.read_csv(f) for f in glob.glob('citibike_data/*.csv')], ignore_index = True)
```

```python
# Display results
clean_data.head()
```

```python
# Get a count of the data in each column to see if they match
clean_data.count()
```

```python
# 
clean_data.gender[clean_data.gender == 0] = 'unknown'
clean_data.gender[clean_data.gender == 1] = 'male'
clean_data.gender[clean_data.gender == 2] = 'female'
clean_data.head()
```

```python
# Export to csv
clean_data.to_csv('clean_data/cleandata2020.csv', mode = 'w', index=False)
```
